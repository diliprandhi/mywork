#10. Adding a new column ‘Trend’ whose values are based on the 'Daily Return'

```python
def trend(x):
  if x > -0.015 and x <= 0.015:
    return 'Slight or No change'
  elif x > 0.015 and x <= 0.04:
    return 'Slight Positive'
  elif x < -0.015 and x >= -0.4:
    return 'Slight Negative'
  elif x > 0.04 and x <= 0.06:
    return 'Positive'
  elif x < -0.04 and x >= -0.06:
    return 'Negative'
  elif x > 0.06 and x <= 0.07:
    return 'Among top gainers'
  elif x < -0.06 and x >= -0.07:
    return 'Among top losers'
  elif x > 0.07:
    return 'Bull run'
  elif x <= -0.07:
    return 'Bear drop'
```

```python
amzn_test['Trend'] = amzn_test['Daily Return'].apply(lambda x: trend(x))
amzn_test.tail(10)
```

```python
aapl_test['Trend'] = aapl_test['Daily Return'].apply(lambda x: trend(x))
aapl_test.tail(10)
```

```python
msft_test['Trend'] = msft_test['Daily Return'].apply(lambda x: trend(x))
msft_test.tail(10)
```

```python
goog_test['Trend'] = goog_test['Daily Return'].apply(lambda x: trend(x))
goog_test.tail(10)
```
